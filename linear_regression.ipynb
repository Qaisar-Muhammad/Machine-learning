```python
import pandas   as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from  sklearn import linear_model
```

```python
df=pd.read_csv('linear_regression_data.csv')
```

```python
df
```

```python
%matplotlib inline 
plt.xlabel=('area(sqr_ft)')
plt.ylabel=('price (US$)')
plt.scatter(df.Area_sqft,df.Price,color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Area_sqft']],df.Price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
