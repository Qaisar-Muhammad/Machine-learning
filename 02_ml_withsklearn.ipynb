```python
import numpy as np 
import pandas as pd 
import matplotlib.pylab as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression
# Load dataset
df = sns.load_dataset('tips')
df.head()
df['smoker_num']=df['smoker'].map({'Yes':1 ,'No': 0 })  # here we did the encoding method   for catagorical data because of machine learing only khows numerical values 
df.head()
# Define features 
x = df[['total_bill','size','smoker_num']]
# this is ower  target coloumn 
y = df['tip']
# Train model
model = LinearRegression()
model.fit(x, y)
# Predict example
print(model.predict([[30,2, 1]]))
print(model.predict([[30,2, 0]]))
```

```python
import numpy as np 
import pandas as pd 
import matplotlib.pylab as plt 
import seaborn as sns
from sklearn.linear_model import LinearRegression ,LogisticRegression
# Load dataset
df = sns.load_dataset('tips')
df.head()
# Define features 
x = df[['total_bill','size','tip']]
# this is ower  target coloumn 
y = df['smoker']
# Train model
model = LogisticRegression()
model.fit(x, y)
# Predict example
print(model.predict([[30.45,2, 5.67]]))
print(model.predict([[10.45,2, 0.46]]))
# here we did logistics regression  ,which gives result in catagorical variable 

```
