```python
# creatind a csv file from that we are doing prediction

from sklearn.linear_model import LinearRegression
import pandas as pd

# Example data
data = {
    'Area_sqft': [800, 950, 1100, 1200, 1500, 1800],
    'Price': [120000, 145000, 170000, 190000, 240000, 290000]
}
df = pd.DataFrame(data)

# Define features and target
x = df[['Area_sqft']]  # 2D array
y = df['Price']        # 1D array

# Create and train model
model = LinearRegression()
model.fit(x, y)  # Fit = Train the model

# Predict example
predicted_price = model.predict([[2000]])
print(predicted_price)
```
